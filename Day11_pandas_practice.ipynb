 Filtering in Pandas means selecting only the rows (or columns) of a DataFrame
  that satisfy certain conditions. It is commonly used to extract relevant data from a larger dataset.

  Syntax: df[condition] -- to find exact rows df[df[condition]]

```python
# filtering demo
df[df['temp_max']>32]
```

Sorting in Pandas means arranging the data in a specific order, either ascending (smallest to largest, A to Z) or descending (largest to smallest, Z to A) based on one or more columns.

It helps organize data and makes analysis easier.

Syntax: df.sort_values(by=['column1', 'column2'])

```python
#sort by mandal
df.sort_values(by='mandal')
```

```python
#sort by different order or columns
df.sort_values(
    by=['mandal', 'district'],
    ascending=[True, False]
)
```

GroupBy in Pandas is used to split the data into groups based on one or more columns and then perform operations such as aggregation, transformation, or filtering on each group.

It follows the Split → Apply → Combine approach:

Split: Divide the data into groups.

Apply: Perform a function on each group.

Combine: Merge the results into a new DataFrame or Series.

1. Functions like count(), size(), first(), last(), and unique() work with non-numeric data.

2. Functions like mean(), sum() (for arithmetic purposes), and median() generally require numeric columns.

```python
df.groupby('district')['rainfall'].mean()
```

Aggregate (or aggregation) in Pandas means applying one or more summary functions to data to produce condensed results. It is commonly used with groupby() to calculate statistics such as sum, mean, maximum, minimum, and count.

The agg() function allows you to apply multiple aggregation functions at once.

Syntax: df.agg(function)

Multiple Aggregations Example: df.groupby('Department')['Salary'].agg(
    ['sum', 'mean', 'max', 'min']
)

```python
#Aggregate function
df[['temp_min','temp_max']].agg(['mean','max'])
print(df)
```

Date: Working with dates involves converting, extracting, filtering, and performing calculations on date/time data.

Convert to datetime: pd.to_datetime()

Extract components: .dt.year, .dt.month, .dt.day

Filter dates: df[df['Date'] > 'YYYY-MM-DD']

Create date ranges: pd.date_range()

Date arithmetic: pd.Timedelta()

**Attribute	Description**

.dt.year	Extract year

.dt.month	Extract month

.dt.day	Extract day

.dt.hour	Extract hour

.dt.minute	Extract minute

.dt.second	Extract second

.dt.day_name()	Name of the weekday

.dt.month_name()	Name of the month

.dt.weekday	Weekday number (Monday=0)


A Pivot Table in Pandas is used to summarize and reorganize data by grouping values and applying aggregation functions. It creates a spreadsheet-like table that helps analyze relationships between variables.

Pivot tables are especially useful for generating reports and comparing data across categories.

Parameters:

data → DataFrame to analyze.

values → Column(s) to aggregate.

index → Row labels (grouping).

columns → Column labels (grouping).

aggfunc → Aggregation function (mean, sum, count, etc.).

fill_value → Value used to replace missing values.


```python
pd.pivot_table(
    df,
    values='column_to_aggregate',
    index='row_grouping_column',
    columns='column_grouping_column',
    aggfunc='aggregation_function'
)
```

**Excercise:**

Create dataset for student Name, age, amrks, deparment, exam date

1. Filtering

Display the details of students who scored more than 80 marks.

2. Sorting

Sort the DataFrame based on Marks in descending order.

3. Adding Columns

Create a new column called Grade using the following criteria:

Marks ≥ 90 → A
Marks between 80 and 89 → B
Marks between 70 and 79 → C
Marks < 70 → D

4. Modifying Columns

Convert all student names into uppercase letters.

5. Removing Columns

Remove the Department column from the DataFrame.

6. Removing Rows

Remove the student record present at index 3.

7. GroupBy

Find the average marks of students in each department.

8. Aggregate Functions

Calculate the following statistics for the Marks column:

Mean
Maximum
Minimum
Sum
Median
Standard Deviation

9. Combining DataFrames

Create another DataFrame containing the following records:

Name	Age	Marks	Department

Teja	20	87	   CSE

Divya	21	91	   ECE

Combine it with the original DataFrame.

10. Working with Dates

Convert the Exam_Date column into datetime format and extract the following:
Day
Month
Year
11. Indexing

Set the Name column as the index and display the details of student "Rahul".

12. Pivot Table

Create a pivot table showing the average Marks for each Department.

```python
import pandas as pd

# -----------------------------------------
# Create Student Dataset
# -----------------------------------------
data = {
    "Name": ["Rahul", "Sneha", "Arjun", "Priya", "Kiran",
             "Anjali", "Vikram", "Pooja", "Rohan", "Meena"],
    "Age": [20, 21, 19, 22, 20, 21, 23, 20, 22, 19],
    "Marks": [85, 92, 78, 65, 88, 73, 95, 81, 69, 90],
    "Department": ["CSE", "ECE", "MECH", "CIVIL", "CSE",
                   "ECE", "CSE", "MECH", "CIVIL", "ECE"],
    "Exam_Date": ["2025-05-10", "2025-05-12", "2025-05-15",
                  "2025-05-18", "2025-05-20", "2025-05-22",
                  "2025-05-25", "2025-05-27", "2025-05-29",
                  "2025-05-30"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# =====================================================
# 1. Filtering
# Display students who scored more than 80 marks
# =====================================================
print("\n1. Students scoring more than 80 marks:")
print(df[df["Marks"] > 80])

# =====================================================
# 2. Sorting
# Sort based on Marks in descending order
# =====================================================
print("\n2. Sorted by Marks (Descending):")
sorted_df = df.sort_values(by="Marks", ascending=False)
print(sorted_df)

# =====================================================
# 3. Adding Columns
# Create Grade column
# =====================================================
def assign_grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 80:
        return "B"
    elif mark >= 70:
        return "C"
    else:
        return "D"

df["Grade"] = df["Marks"].apply(assign_grade)

print("\n3. DataFrame with Grade Column:")
print(df)

# =====================================================
# 4. Modifying Columns
# Convert names to uppercase
# =====================================================
df["Name"] = df["Name"].str.upper()

print("\n4. Names converted to uppercase:")
print(df)

# =====================================================
# 5. Removing Columns
# Remove Department column
# =====================================================
df_without_dept = df.drop("Department", axis=1)

print("\n5. DataFrame after removing Department:")
print(df_without_dept)

# =====================================================
# 6. Removing Rows
# Remove student record at index 3
# =====================================================
df_removed_row = df.drop(index=3)

print("\n6. DataFrame after removing row at index 3:")
print(df_removed_row)

# =====================================================
# 7. GroupBy
# Average marks by department
# =====================================================
print("\n7. Average Marks by Department:")
avg_marks = df.groupby("Department")["Marks"].mean()
print(avg_marks)

# =====================================================
# 8. Aggregate Functions
# =====================================================
print("\n8. Aggregate Statistics for Marks:")

print("Mean:", df["Marks"].mean())
print("Maximum:", df["Marks"].max())
print("Minimum:", df["Marks"].min())
print("Sum:", df["Marks"].sum())
print("Median:", df["Marks"].median())
print("Standard Deviation:", df["Marks"].std())

# =====================================================
# 9. Combining DataFrames
# =====================================================
new_data = {
    "Name": ["Teja", "Divya"],
    "Age": [20, 21],
    "Marks": [87, 91],
    "Department": ["CSE", "ECE"]
}

new_df = pd.DataFrame(new_data)

# Match existing columns
new_df["Exam_Date"] = pd.NaT
new_df["Grade"] = new_df["Marks"].apply(assign_grade)

combined_df = pd.concat([df, new_df], ignore_index=True)

print("\n9. Combined DataFrame:")
print(combined_df)

# =====================================================
# 10. Working with Dates
# =====================================================
df["Exam_Date"] = pd.to_datetime(df["Exam_Date"])

df["Day"] = df["Exam_Date"].dt.day
df["Month"] = df["Exam_Date"].dt.month
df["Year"] = df["Exam_Date"].dt.year

print("\n10. Extracted Day, Month, Year:")
print(df[["Exam_Date", "Day", "Month", "Year"]])

# =====================================================
# 11. Indexing
# Set Name as index and display Rahul's details
# =====================================================
indexed_df = df.set_index("Name")

print("\n11. Details of student RAHUL:")
print(indexed_df.loc["RAHUL"])

# =====================================================
# 12. Pivot Table
# Average Marks for each Department
# =====================================================
pivot = pd.pivot_table(
    df,
    values="Marks",
    index="Department",
    aggfunc="mean"
)

print("\n12. Pivot Table (Average Marks by Department):")
print(pivot)
```

**What is Matplotlib?**

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

Install and import matplotlib module

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

*import matplotlib.pyplot as plt

Let's create one Dataset (Student with feilds - Name, Marks, Hours, Department).

Let's Use ONE Dataset for ALL Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
      "Name":["Joyce","Hema","Kavya","Harsha","Bobby"],
      "Marks":[60,79,50,85,76],
      "Hours":[3,4,2,5,4],
      "Department":["CSE","CSE","IT","AIML","ECE"]
}
df=pd.DataFrame(data)
print(df)
```

**Anatomy of Any Graph**

Every graph has:

Title

X-axis

Y-axis

Data

Legend (optional)

The following plots are considered the most important and frequently used in assignments, projects, data analysis, machine learning, and research work.

1. Line Plot (plt.plot())

Purpose: Shows trends or changes over time.

When to Use?

Temperature changes

Sensor readings

Stock prices

Semester GPA trends

Common Example: Marks vs Study Hours

```python
plt.plot(df['Name'], df['Marks'])
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()

#Use Line Plot when order matters.
```

2. Bar Chart(plt.bar())

Purpose: Compare quantities among categories.

When to use?

Department placements

Product sales

Machine efficiency

Example: Department-wise count.

```python
dept_count = df['Department'].value_counts()

plt.bar(dept_count.index, dept_count.values)
plt.show()

#Use Bar Chart for comparisons.
```

3. Scatter Plot

Purpose: Displays relationships between two variables.

Used in:

Machine Learning

Regression analysis

Experimental data analysis

```python
plt.scatter(df['Hours'], df['Marks'])
plt.show()

'''Scatter plots answer:
"If X changes, what happens to Y?, Relationship between study time and marks.'''
```

4. Histogram (plt.hist())

Purpose: Shows frequency distribution of data.

Used in:

Statistics

Probability

Data Science

Quality control

```python
plt.hist(df['Marks'])
plt.show()
```

you are giving only one variable (df['Marks']) as input. In a histogram, Matplotlib automatically decides the axes as follows:

X-axis: The ranges (bins) of the data values.
Y-axis: The frequency (count) of observations that fall into each bin.

Marks = [35, 40, 42, 45, 48, 52, 55, 58, 60, 65]

30-40 -->         1

5. Box Plot: (plt.boxplot())

Purpose: Identifies spread, median, and outliers.

Used in:

1. Statistical analysis

2. Experimental validation

3. Data preprocessing






```python
plt.boxplot(df['Marks'])
plt.show()
```

Y-axis: The actual data values (Marks).

X-axis: The position/index of the box (not your data values).

For a single box plot, the X-axis usually just shows 1 because there is only one dataset.

Above box plot tells us that:

The lowest value is 50.

The highest value is 80.

The median value is 60.

The middle 50% of the data lies between 55 and 70.

There are no outliers.

The data appears slightly more spread above the median (60 to 80) than below it (50 to 60).

7. Heatmap (plt.imshow())

A heatmap is a graphical representation of data where the values of a matrix are represented using different colors. It helps you quickly identify patterns, high and low values, and correlations.

Heatmaps are widely used in:

🖼️ Image Processing – visualizing grayscale images and pixel intensities.

🤖 Machine Learning – displaying confusion matrices and correlation matrices.

📡 Signal Processing – showing spectrograms and frequency distributions.

⚡ Electrical Engineering – power consumption and sensor data analysis.

🏭 Industrial Engineering – process monitoring and quality control.

```python
import pandas as pd

data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Hours': [2, 4, 5, 7, 8],
    'Marks': [60, 70, 75, 85, 90],
    'Attendance': [70, 80, 85, 90, 95]
}

df = pd.DataFrame(data)

print(df)
```

```python
corr = df[['Hours', 'Marks', 'Attendance']].corr()

print(corr)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))

plt.imshow(corr)

plt.colorbar(label='Correlation')

plt.xticks(
    range(len(corr.columns)),
    corr.columns
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Student Dataset Correlation Heatmap")

plt.show()
```

+1 → Strong positive relationship

0 → No relationship

−1 → Strong negative relationship

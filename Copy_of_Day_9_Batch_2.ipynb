** Cleaning Data **

Data cleaning means fixing bad data in your data set.

Bad data could be:

1. Empty cells

2. Data in wrong format

3. Wrong data

4. Duplicates



One way to deal with empty cells is to remove rows that contain empty cells.



```python
import pandas as pd

df = pd.read_csv('studentdata.csv')

new_df = df.dropna()

print(new_df.to_string())
```

```python
import pandas as pd

# Step 1: Create Student Dataset
data = {
    "StudentID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Rahul", "Priya", "Amit", "Sneha", "Kiran",
             "Pooja", "Arjun", "Divya", "Vikram", "Neha"],
    "Age": [15, 16, 15, 17, 16, 15, 17, 16, 15, 17],
    "Gender": ["Male", "Female", "Male", "Female", "Male",
               "Female", "Male", "Female", "Male", "Female"],
    "Class": ["10A", "10A", "10B", "10B", "10A",
              "10B", "10A", "10B", "10A", "10B"],
    "Marks": [92, 88, 76, 95, 45, 82, 67, 91, 39, 85],
    "City": ["Hyderabad", "Mumbai", "Hyderabad", "Delhi", "Chennai",
             "Hyderabad", "Mumbai", "Delhi", "Hyderabad", "Chennai"],
    "Email": [
        "rahul@gmail.com", "priya@gmail.com", "amit@gmail.com",
        "sneha@gmail.com", "kiran@gmail.com", "pooja@gmail.com",
        "arjun@gmail.com", "divya@gmail.com", "vikram@gmail.com",
        "neha@gmail.com"
    ]
}

# Step 2: Load Dataset
df = pd.DataFrame(data)

# Step 3: Display first 5 rows
print("First 5 Rows:")
print(df.head())

# Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail())

# Show column names
print("\nColumn Names:")
print(df.columns)

# Dataset info
print("\nDataset Info:")
print(df.info())

# Students with marks > 85
print("\nStudents with Marks > 85:")
print(df[df["Marks"] > 85])

# Students from Hyderabad
print("\nStudents from Hyderabad:")
print(df[df["City"] == "Hyderabad"])

# Female students
print("\nFemale Students:")
print(df[df["Gender"] == "Female"])

# Which city has the highest average marks?
city_avg = df.groupby("City")["Marks"].mean()
print("\nAverage Marks by City:")
print(city_avg)

highest_city = city_avg.idxmax()
print("\nCity with Highest Average Marks:", highest_city)

# Who is the topper in each class?
topper_each_class = df.loc[df.groupby("Class")["Marks"].idxmax()]
print("\nTopper in Each Class:")
print(topper_each_class[["Class", "Name", "Marks"]])

# How many students failed? (Pass mark = 40)
failed_students = df[df["Marks"] < 40]
print("\nNumber of Failed Students:", len(failed_students))

# Overall pass percentage
total_students = len(df)
passed_students = len(df[df["Marks"] >= 40])

pass_percentage = (passed_students / total_students) * 100
print("\nOverall Pass Percentage:", round(pass_percentage, 2), "%")
```

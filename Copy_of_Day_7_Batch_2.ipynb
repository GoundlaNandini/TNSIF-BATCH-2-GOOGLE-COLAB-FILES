**Trees, Graph and Hash table Implementation**

```python
# Node class for Binary Search Tree
class BSTNode:
    def __init__(self, key):
        # Store the value of the node
        self.key = key

        # Pointer to the left child
        self.left = None

        # Pointer to the right child
        self.right = None


# Binary Search Tree class
class BST:
    def __init__(self):
        # Initially, the tree is empty
        self.root = None

    # Insert a new key into the BST
    def insert(self, root, key):
        # If the current position is empty, create a new node
        if root is None:
            return BSTNode(key)

        # If key is smaller, insert into the left subtree
        if key < root.key:
            root.left = self.insert(root.left, key)

        # If key is larger, insert into the right subtree
        elif key > root.key:
            root.right = self.insert(root.right, key)

        # Return the updated root node
        return root

    # Search for a key in the BST
    def search(self, root, key):
        # Return the node if found, or None if not found
        if root is None or root.key == key:
            return root

        # Search in the left subtree if key is smaller
        if key < root.key:
            return self.search(root.left, key)

        # Otherwise, search in the right subtree
        return self.search(root.right, key)

    # Perform inorder traversal (Left → Root → Right)
    def inorder(self, root):
        if root:
            # Visit left subtree
            self.inorder(root.left)

            # Print current node
            print(root.key, end=" ")

            # Visit right subtree
            self.inorder(root.right)
```

**AVL** Tree

```python
# Node class for AVL Tree
class AVLNode:
    def __init__(self, key):
        # Value stored in the node
        self.key = key

        # Pointer to left child
        self.left = None

        # Pointer to right child
        self.right = None

        # Height of the node (used for balancing)
        self.height = 1


# AVL Tree class
class AVLTree:

    # Get height of a node (0 if node is None)
    def height(self, node):
        return node.height if node else 0

    # Compute balance factor of a node
    # balance = height(left subtree) - height(right subtree)
    def balance(self, node):
        if not node:
            return 0
        return self.height(node.left) - self.height(node.right)

    # Right rotation (used for Left-Left imbalance)
    def right_rotate(self, y):
        # y is the unbalanced node
        x = y.left
        t2 = x.right

        # Perform rotation
        x.right = y
        y.left = t2

        # Update heights after rotation
        y.height = 1 + max(self.height(y.left),
                           self.height(y.right))

        x.height = 1 + max(self.height(x.left),
                           self.height(x.right))

        # New root after rotation
        return x

    # Left rotation (used for Right-Right imbalance)
    def left_rotate(self, x):
        # x is the unbalanced node
        y = x.right
        t2 = y.left

        # Perform rotation
        y.left = x
        x.right = t2

        # Update heights after rotation
        x.height = 1 + max(self.height(x.left),
                           self.height(x.right))

        y.height = 1 + max(self.height(y.left),
                           self.height(y.right))

        # New root after rotation
        return y

    # Insert a key into the AVL tree and rebalance if needed
    def insert(self, root, key):

        # Step 1: Perform normal BST insertion
        if not root:
            return AVLNode(key)

        if key < root.key:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        # Step 2: Update height of current node
        root.height = 1 + max(
            self.height(root.left),
            self.height(root.right)
        )

        # Step 3: Get balance factor to check if node became unbalanced
        balance = self.balance(root)

        # -------------------------
        # Step 4: Handle 4 cases of imbalance
        # -------------------------

        # Left Left (LL) case
        if balance > 1 and key < root.left.key:
            return self.right_rotate(root)

        # Right Right (RR) case
        if balance < -1 and key > root.right.key:
            return self.left_rotate(root)

        # Left Right (LR) case
        if balance > 1 and key > root.left.key:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        # Right Left (RL) case
        if balance < -1 and key < root.right.key:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        # If balanced, return unchanged node
        return root
```

```python
# Color constants used in Red-Black Tree
RED = True
BLACK = False


# Node class for Red-Black Tree
class RBNode:
    def __init__(self, key):
        # Value stored in the node
        self.key = key

        # New nodes are always inserted as RED initially
        self.color = RED

        # Pointer to left child
        self.left = None

        # Pointer to right child
        self.right = None

        # Pointer to parent node (needed for fix-up operations)
        self.parent = None
```

```python
# Left rotation in Red-Black Tree
def left_rotate(root, x):
    # Set y as x's right child (pivot node for rotation)
    y = x.right

    # Turn y's left subtree into x's right subtree
    x.right = y.left

    # If y's left subtree exists, update its parent to x
    if y.left:
        y.left.parent = x

    # Link y's parent to x's parent (moving y up the tree)
    y.parent = x.parent

    # If x was the root node, update root to y
    if x.parent is None:
        root = y

    # If x was a left child, update its parent's left pointer
    elif x == x.parent.left:
        x.parent.left = y

    # Otherwise, x was a right child
    else:
        x.parent.right = y

    # Put x as left child of y
    y.left = x

    # Update x's parent pointer
    x.parent = y

    # Return new root of the tree
    return root
```

**Graph Implementation**

```python
class Graph:
    def __init__(self):
        # Dictionary to store graph:
        # key = node, value = list of neighbors
        self.graph = {}

    # Add a node
    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    # Add an edge (undirected graph)
    def add_edge(self, u, v):
        # If nodes don't exist, create them
        if u not in self.graph:
            self.add_vertex(u)
        if v not in self.graph:
            self.add_vertex(v)

        # Add connection both ways
        self.graph[u].append(v)
        self.graph[v].append(u)

    # Display graph
    def display(self):
        for node, neighbors in self.graph.items():
            print(f"{node} -> {neighbors}")
```

```python
g = Graph()

g.add_edge("A", "B")
g.add_edge("A", "C")
g.add_edge("B", "D")
g.add_edge("C", "D")

g.display()
```

```python
# BFS Traversal (Graph)

from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            for neighbor in graph[node]:
                queue.append(neighbor)


# Run BFS
bfs(g.graph, "A")
```
